```python
import scipy.io as sio
import numpy as np
import os
import csv
import pandas as pd
os.chdir('/home/zohyos7/fmri')
```

```python
mat_contents = sio.loadmat('youngDat_N250_Dec2015.mat')
```

```python
mat_contents['r'][0,0].dtype.descr
```

```python
label = []
for i in range (mat_contents['r'][0,0]['politicals'][0].shape[0]):
    val = mat_contents['r'][0,0]['politicals'][0][i][0]
    label.append(val)
political_label = dict((i+1,label[i]) for i in range(len(label)))
```

```python
bad_subject_list = []
for i in range (mat_contents['r'][0,0]['badSubjectIDs'].shape[0]):
    val = mat_contents['r'][0,0]['badSubjectIDs'][i][0]
    bad_subject_list.append(val)
bad_subject_list
```

```python
political_label
```

```python
label = []
for i in range (mat_contents['r'][0,0]['languages'][0].shape[0]):
    val = mat_contents['r'][0,0]['languages'][0][i][0]
    label.append(val)
language_label = dict((i+1,label[i]) for i in range(len(label))) 
```

```python
language_label
```

```python
languagelist = []
for i in range (mat_contents['r'][0,0]['languageIndex'].shape[0]):
    val = mat_contents['r'][0,0]['languageIndex'][i][0][0][0]
    languagelist.append(val)
```

```python
subjID = pd.DataFrame(mat_contents['r'][0,0]['subjID'], columns = ['subjID'])
```

```python
political = pd.DataFrame(mat_contents['r'][0,0]['political'], columns = ['political affiliation'])
```

```python
language = pd.DataFrame(languagelist, columns = ['language'])
```

```python
birthPlace = pd.DataFrame(mat_contents['r'][0,0]['birthPlaceUSA'], columns = ['birthPlace'])
```

```python
bad_subject = pd.DataFrame(mat_contents['r'][0,0]['isBadSubjects'].T, columns = ['isBadSubjects'])
```

```python
subj_profile = pd.concat([subjID, political, language, birthPlace, bad_subject], axis=1)
```

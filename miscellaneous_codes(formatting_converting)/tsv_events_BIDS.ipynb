```python
for filename in os.listdir("."):
    for functask in func_task:
        if os.path.exists(os.path.join(filename,'func')) == True:
            a = pd.read_csv(os.path.join(filename,'func',"{}{}{}{}".format(filename, '_task-', functask, '_bold.tsv')),delimiter='\t',encoding='utf-8')
            a = a.drop(a.columns[0], axis =1)
            a['duration'] = 2.0
            a = a.rename(columns = {'startSecs':'onset'})
            cols = a.columns.tolist()
            cols = cols[-2:] + cols[:-2]
            a = a[cols]
            a = a.fillna('n/a')
            a.to_csv(os.path.join(filename,'func',"{}{}{}{}".format(filename,'_task-', functask, '_events.tsv')),
                sep = '\t', quoting=csv.QUOTE_NONE, index = False)
            print(filename, functask)
```
